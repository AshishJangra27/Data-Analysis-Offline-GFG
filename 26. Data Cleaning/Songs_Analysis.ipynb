```python
# !git clone https://github.com/AshishJangra27/Gaana-Data-Analysis;

import pandas as pd

df = pd.read_csv('/content/Gaana-Data-Analysis/songs.csv')
df.drop(index=[36814], axis=0, inplace=True)
df.head()
```

### 1. Cleaning the duration

```python
clean_duration = []

for i in df['duration']:
  if len(i.split(':')) == 2:       # Min:Sec
    clean_duration.append(int(i.split(':')[0])*60  + int(i.split(':')[1]))
  else:
    clean_duration.append(int(i.split(':')[0])*60*60 + int(i.split(':')[1])*60 + int(i.split(':')[2]))

df['clean_duration'] = clean_duration

df.head()
```

### 2. Exlporing Languages

```python
df['language'].nunique()
```

### 3. Cleaning singer_id

```python
df['singer_id'] = [i.replace('/artist/','').replace('|',',') for i in df['singer_id']]
df.head()
```

### 4. Finding songs in each language

```python
for lang in df['language'].unique():
  freq = 0
  for i in df['language']:
    if i == lang:
      freq += 1
  print(lang , freq)
```

### 5. Finding songs in each language using sub-dataframe

```python
for lang in df['language'].unique():
  print(lang, len(df[df['language'] == lang]))
```
