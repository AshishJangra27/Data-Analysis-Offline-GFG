#### 1.2) Loading the Dataset

##### 1.6) Removing Duplicates

```python
df.drop_duplicates(subset='link', inplace=True)
```

### 2. Data Cleaning

#### 2.1) Cleaning Duration

```python
duration = []

for i in df['duration']:

  mints = int(i.split(":")[0])
  secds = int(i.split(":")[1])

  duration.append(mints*60 + secds)

df['duration_in_seconds'] = duration

df.head()
```
