### 1. Cleaning the duration

```python
clean_duration = []

for i in df['duration']:
  if len(i.split(':')) == 2:       # Min:Sec
    clean_duration.append(int(i.split(':')[0])*60  + int(i.split(':')[1]))
  else:
    clean_duration.append(int(i.split(':')[0])*60*60 + int(i.split(':')[1])*60 + int(i.split(':')[2]))

df['clean_duration'] = clean_duration
del df['duration']

df.head()
```
