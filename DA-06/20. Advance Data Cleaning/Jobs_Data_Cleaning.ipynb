```python
import pandas as pd

df = pd.read_csv('/content/datasets/Job Postings/jobs.csv.zip')
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df['reviews'] = df['reviews'].str.split(' ').str[0].astype(int)

df
```

```python
df['exp_start'] = df['experience'].str.replace(' Yrs','').str.split('-').str[0].astype(int)
df['exp_end']   = df['experience'].str.replace(' Yrs','').str.split('-').str[1].astype(int)

df.head()
```

```python
x = []
y = []

for yrs in range(df['exp_end'].max()+1):

  indx = 0
  indexes = []

  for i in df['experience']:
    st,end = i.replace(' Yrs','').split('-')
    if (yrs >= int(st) and yrs <= int(end) ):
      indexes.append(indx)

    indx += 1

  x.append(yrs)
  y.append(len(df.iloc[indexes]))
```

```python
import re

df['salary'] = df['salary'].str.replace(' Cr','00,00,000')
df['salary'] = df['salary'].str.split('(').str[0]

results = []

for i in df['salary'].astype(str):

  i = i.replace('.',',')

  res = re.findall(r"\d+(?:,\d+)*", i)

  if (len(res) == 1):
      results.append([res[0].replace(',',''),res[0].replace(',','')])

  elif (len(res) == 2):
    results.append([res[0].replace(',',''),res[1].replace(',','')])

  elif (len(res) == 0):
    results.append(['NA','NA'])


df['salary_min'] = [i[0] for i in results]
df['salary_max'] = [i[1] for i in results]

df.head()
```

```python
responsibilities = []

for i in df['resposibilities']:
  responsibilities += i.split(',')

responsibilities = list((set(responsibilities)))
```

```python
from tqdm import tqdm

data = []

for responsibility in tqdm(responsibilities):
  c = 0

  for i in df['resposibilities']:
    if responsibility.lower() in i.lower().split(','):
      c += 1

  data.append([responsibility,c])
```

```python
df = pd.read_csv('/content/datasets/Myntra Products/products.csv')
df.head()
```

```python
df.groupby('brand_name')['discounted_price'].max().sort_values(ascending=False).head(10).plot(kind = 'bar')
```

```python
import matplotlib.pyplot as plt

# Group data and get top 10 brands by maximum discounted price
brand_prices = df.groupby('brand_name')['discounted_price'].max().sort_values(ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
bars = plt.bar(brand_prices.index, brand_prices.values, color='skyblue', width=0.6)

# Customizations
plt.xlabel("Brand Name", fontsize=12)
plt.ylabel("Maximum Discounted Price", fontsize=12)
plt.title("Top 10 Brands by Maximum Discounted Price", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels
plt.yticks(fontsize=10)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)  # Add gridlines

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, round(yval, 2), ha='center', va='bottom', fontsize=8)

# Display the plot
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
